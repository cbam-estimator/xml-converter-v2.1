```python
from reportlab.lib import colors
from reportlab.lib.pagesizes import A4
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib.units import inch, cm
from reportlab.platypus import SimpleDocTemplate, Paragraph, Table, TableStyle, Spacer

def create_pdf(filename):
    # Erstelle ein PDF-Dokument mit A4-Größe
    doc = SimpleDocTemplate(filename, pagesize=A4,
                            rightMargin=2*cm, leftMargin=2*cm,
                            topMargin=2*cm, bottomMargin=2*cm)

    # Elemente, die dem PDF hinzugefügt werden
    elements = []

    # Stile für den Titel und den Text
    styles = getSampleStyleSheet()
    title_style = styles['Title']
    title_style.alignment = 1  # Zentriert

    subtitle_style = styles['Heading2']
    subtitle_style.alignment = 1  # Zentriert

    # Titel hinzufügen
    title = Paragraph("CBAM Estimator", title_style)
    elements.append(title)
    elements.append(Spacer(1, 0.5*cm))  # Abstand

    # Untertitel hinzufügen
    subtitle = Paragraph("Kontaktversuche:", subtitle_style)
    elements.append(subtitle)
    elements.append(Spacer(1, 0.3*cm))  # Abstand

    # Daten für die Tabelle (Beispiel)
    data = [
        ['Name', 'Datum', 'Status'],
        ['Max Mustermann', '2024-04-01', 'Erfolgreich'],
        ['Erika Musterfrau', '2024-04-03', 'Kein Antwort'],
        ['John Doe', '2024-04-05', 'Erfolgreich'],
        ['Jane Smith', '2024-04-07', 'Ausstehend']
    ]

    # Tabelle erstellen
    table = Table(data, colWidths=[6*cm, 4*cm, 4*cm])

    # Stil für die Tabelle
    style = TableStyle([
        ('BACKGROUND', (0,0), (-1,0), colors.grey),
        ('TEXTCOLOR',(0,0),(-1,0),colors.whitesmoke),

        ('ALIGN',(0,0),(-1,-1),'CENTER'),
        ('FONTNAME', (0,0), (-1,0), 'Helvetica-Bold'),
        ('FONTSIZE', (0,0), (-1,0), 12),

        ('BOTTOMPADDING', (0,0), (-1,0), 12),
        ('BACKGROUND',(0,1),(-1,-1),colors.beige),

        ('GRID', (0,0), (-1,-1), 1, colors.black),
    ])
    table.setStyle(style)

    elements.append(table)

    # PDF generieren
    doc.build(elements)
    print(f"PDF '{filename}' wurde erfolgreich erstellt.")

if __name__ == "__main__":
    create_pdf("cbam_estimator.pdf")
```

```python
import os
from reportlab.lib import colors
from reportlab.lib.pagesizes import A4
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib.units import cm
from reportlab.platypus import SimpleDocTemplate, Paragraph, Table, TableStyle, Spacer

# Erstelle das Verzeichnis, falls es nicht existiert
output_dir = os.path.join("..", "output")
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

def create_pdf(filename, entrielist):
    # Pfad für die PDF-Datei im output-Verzeichnis
    filepath = os.path.join(output_dir, filename)

    # Erstelle ein PDF-Dokument mit A4-Größe
    doc = SimpleDocTemplate(filepath, pagesize=A4,
                            rightMargin=2*cm, leftMargin=2*cm,
                            topMargin=2*cm, bottomMargin=2*cm)

    # Elemente, die dem PDF hinzugefügt werden
    elements = []

    # Stile für den Titel und den Text
    styles = getSampleStyleSheet()
    title_style = styles['Title']
    title_style.alignment = 1  # Zentriert

    # Titel hinzufügen
    title = Paragraph("Dokumentation gescheiterter Kontaktversuche zur Ermittlung der Emissionsdaten", title_style)
    elements.append(title)
    elements.append(Spacer(1, 0.5*cm))  # Abstand

    # Tabellen-Daten aufbereiten
    table_data = [['Kommunikationsversuch #', 'Datum', 'Anmerkungen']]  # Tabellenkopf

    # Einträge der entrielist in die Tabelle einfügen
    for idx, entry in enumerate(entrielist, start=1):
        # Anmerkungen aus remarks_dropdown und remarks kombinieren
        anmerkungen = f"{entry['remarks_dropdown']} - {entry['remarks']}"
        # Füge die Zeile hinzu
        table_data.append([str(idx), entry['date'], anmerkungen])

    # Tabelle erstellen
    table = Table(table_data, colWidths=[5*cm, 4*cm, 7*cm])

    # Stil für die Tabelle
    style = TableStyle([
        ('BACKGROUND', (0, 0), (-1, 0), colors.grey),
        ('TEXTCOLOR', (0, 0), (-1, 0), colors.whitesmoke),

        ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
        ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
        ('FONTSIZE', (0, 0), (-1, 0), 12),

        ('BOTTOMPADDING', (0, 0), (-1, 0), 12),
        ('BACKGROUND', (0, 1), (-1, -1), colors.beige),

        ('GRID', (0, 0), (-1, -1), 1, colors.black),
    ])
    table.setStyle(style)

    elements.append(table)

    # PDF generieren
    doc.build(elements)
    print(f"PDF '{filename}' wurde erfolgreich erstellt im Verzeichnis '{output_dir}'.")

# Beispiel für die entrielist
entrielist = [
    {"date": "2024-10-10", "remarks": "Keine Antwort", "remarks_dropdown": "Versuch 1"},
    {"date": "2024-10-12", "remarks": "Kein Rückruf", "remarks_dropdown": "Versuch 2"},
    {"date": "2024-10-14", "remarks": "Mailbox", "remarks_dropdown": "Versuch 3"}
]

if __name__ == "__main__":
    # PDF erstellen mit Beispieldaten
    create_pdf("kontaktversuche.pdf", entrielist)
```
