```python
import logging
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select, WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Logging konfigurieren
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# Pfad zum ChromeDriver
chrome_service = Service('/opt/homebrew/bin/chromedriver')

# Chrome-Optionen konfigurieren
options = webdriver.ChromeOptions()
options.binary_location = '/Users/andreas/Downloads/chrome-mac-arm64/Google Chrome for Testing.app/Contents/MacOS/Google Chrome for Testing'

# Initialisiere den WebDriver (hier für Chrome)
driver = webdriver.Chrome(service=chrome_service, options=options)

try:
    logging.info('Öffne die Webseite...')
    # Öffne die Webseite
    driver.get('https://cbam.ec.europa.eu/declarant')

    # Warte, bis das erste Dropdown-Menü geladen ist
    logging.info('Warte, bis das erste Dropdown-Menü geladen ist...')
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.NAME, 'domain'))
    )

    # Wähle "Zoll" im ersten Dropdown-Menü aus
    logging.info('Wähle "Zoll" im ersten Dropdown-Menü aus...')
    select_domain = Select(driver.find_element(By.NAME, 'domain'))
    select_domain.select_by_value('CUST')

    # Warte, bis das zweite Dropdown-Menü geladen ist
    logging.info('Warte, bis das zweite Dropdown-Menü geladen ist...')
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.NAME, 'subdomain'))
    )

    time.sleep(30)

    # Wähle "Germany" im zweiten Dropdown-Menü aus
    logging.info('Wähle "Germany" im zweiten Dropdown-Menü aus...')
    select_subdomain = Select(driver.find_element(By.NAME, 'subdomain'))
    select_subdomain.select_by_visible_text('Germany')

    

    # Warte, bis das dritte Dropdown-Menü geladen ist
    logging.info('Warte, bis das dritte Dropdown-Menü geladen ist...')
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.NAME, 'typeOfActor'))
    )

    # Wähle "Wirtschaftsbeteiligter" im dritten Dropdown-Menü aus
    logging.info('Wähle "Wirtschaftsbeteiligter" im dritten Dropdown-Menü aus...')
    select_type_of_actor = Select(driver.find_element(By.NAME, 'typeOfActor'))
    select_type_of_actor.select_by_visible_text('Wirtschaftsbeteiligter')

    # Checke die "mir selbst" Radiobox
    logging.info('Checke die "mir selbst" Radiobox...')
    radio_self = driver.find_element(By.ID, 'optionsRadios1')
    radio_self.click()

    # Checke die Einverständnis-Checkbox
    logging.info('Checke die Einverständnis-Checkbox...')
    consent_checkbox = driver.find_element(By.NAME, 'consent')
    consent_checkbox.click()

    # Drücke den "Abschicken"-Button
    logging.info('Drücke den "Abschicken"-Button...')
    submit_button = driver.find_element(By.CSS_SELECTOR, 'button[type="submit"]')
    submit_button.click()

    # Überprüfen, ob die Weiterleitung erfolgreich war
    logging.info('Überprüfen, ob die Weiterleitung erfolgreich war...')
    WebDriverWait(driver, 10).until(
        EC.url_changes(driver.current_url)
    )
    logging.info('Weiterleitung erfolgreich!')

except Exception as e:
    logging.error('Ein Fehler ist aufgetreten: %s', e)

finally:
    logging.info('Schließe den Browser...')
    # Schließe den Browser
    driver.quit()
```

```python
import tkinter as tk
from tkinter import ttk
from tkinter import scrolledtext
from tkinterdnd2 import TkinterDnD, DND_FILES

class FileDropApp(TkinterDnD.Tk):
    def __init__(self):
        super().__init__()

        # Fenster konfigurieren
        self.title("Drag and Drop App")
        self.geometry("600x400")

        # Dateiablagefläche erstellen
        self.drop_frame = ttk.Frame(self, width=500, height=100, relief=tk.RAISED)
        self.drop_frame.pack(pady=20)
        self.drop_frame.pack_propagate(False)

        self.drop_label = ttk.Label(self.drop_frame, text="Dateien hierher ziehen", anchor="center")
        self.drop_label.pack(expand=True, fill=tk.BOTH)

        # Drag & Drop aktivieren
        self.drop_frame.drop_target_register(DND_FILES)
        self.drop_frame.dnd_bind('<<Drop>>', self.on_drop)

        # Scrollbares Textfeld für Logs erstellen
        self.log_text = scrolledtext.ScrolledText(self, wrap=tk.WORD, height=15, state='disabled')
        self.log_text.pack(fill=tk.BOTH, expand=True, padx=10, pady=10)

    def on_drop(self, event):
        # Log-Nachricht hinzufügen
        file_path = event.data
        self.log_message(f"Datei empfangen: {file_path}", "blue")

    def log_message(self, message, colour="black"):
        # Logs farbig anzeigen
        self.log_text.config(state='normal')
        self.log_text.insert(tk.END, message + '\n', ('colour',))
        self.log_text.tag_config('colour', foreground=colour)
        self.log_text.config(state='disabled')
        self.log_text.yview(tk.END)  # Scroll automatisch ans Ende

if __name__ == "__main__":
    app = FileDropApp()
    app.mainloop()
```

```python

import unicodedata

def non_exact_search(target, search_list, warn=False, warn_title=None):
    # Searches for a non-exact match of a target string in a list of strings
    for search_str in search_list:
        target = normalize_str(target)
        search_str = normalize_str(search_str)
        if clean_supplier_str(target.lower()) == clean_supplier_str(search_str.lower()):
            if warn:
                print("warning non exact match")
            return search_str

    return None

def normalize_str(s):
    return unicodedata.normalize("NFC", s)

def exact_search(target, search_list):
    # Searches for an exact match of a target string in a list of strings

    for search_str in search_list:
        target = normalize_str(target)
        search_str = normalize_str(search_str)
        if target.strip().lower() == search_str.strip().lower():
            return search_str
        else:
            print(f"'{target.strip().lower()}' != '{search_str.strip().lower()}'")

    return None


def clean_supplier_str(supplier_str):
    # Cleans installation or operator string for comparison within supplier tool

    if supplier_str is None:
        return None

    ignore_ls = (
        " ",
        " ",
        "-",
        "_",
        "(",
        ")",
        ".",
        ",",
        ";",
        ":",
        "/",
        "\\",
        "co",
        "ltd",
        "gmbh",
    )

    for word in ignore_ls:
        supplier_str = supplier_str.replace(word, "")

    return supplier_str

search_list = ["MÄDLER  GmbH", "Hello World!", "Hello World!!", "Hello World!!!"]

print(non_exact_search("MÄDLER GmbH", search_list))
# print(non_exact_search("MÄDLER GmbH", search_list))
```

```python
from pdfrw import PdfReader, PdfWriter, PdfDict, PdfName, PdfObject

# Path to your PDF form
pdf_template_path = "/Users/andreas/Lokal/CBAM - Lokal/Workspaces/ce_xml_converter/tests/vorlage_2_ak.pdf"
output_pdf_path = "/Users/andreas/Lokal/CBAM - Lokal/Workspaces/ce_xml_converter/tests/output.pdf"

# Example data to fill in the form
form_data = {
    "deklarant_name": "Musterfirma",
    "deklarant_EORI": "123456789",
    "deklarant_kontaktperson": "Max Mustermann",
    "deklarant_kontaktperson_mail": "max@musterfirma.de",
    "hersteller_name": "Hersteller XYZ",
    "hersteller_land": "Deutschland",
    "hersteller_kontaktperson": "Kontaktperson XYZ",
    "hersteller_kontaktperson_mail": "kontakt@xyz.de",
    "cn_codes": "CN001, CN002",
    "Datum": "2024-10-24",
    "Grund des Scheiterns": "Keine Rückmeldung",
    "Anmerkungen": "Mehrfache Kontaktaufnahme versucht",
    "signature_name": "Max Mustermann",
    "date": "2024-10-24"
}

# Function to fill the PDF form
def fill_pdf(input_pdf_path, output_pdf_path, data_dict):
    # Load the PDF form
    template_pdf = PdfReader(input_pdf_path)

    # Ensure that NeedAppearances is set to true
    if template_pdf.Root.AcroForm:
        template_pdf.Root.AcroForm.update(
            PdfDict(NeedAppearances=PdfObject('true'))
        )
    else:
        template_pdf.Root.AcroForm = PdfDict(
            NeedAppearances=PdfObject('true')
        )

    # Loop through each page
    for page in template_pdf.pages:
        annotations = page['/Annots']
        if annotations:
            # Loop through each annotation
            for annotation in annotations:
                if annotation['/Subtype'] == '/Widget' and annotation['/T']:
                    # Access the field directly without getObject()
                    field = annotation
                    # Extract the field name
                    field_name = field['/T'][1:-1]
                    if field_name in data_dict:
                        # Update the field value
                        field.update(
                            PdfDict(V='{}'.format(data_dict[field_name]))
                        )
                        # Set the field appearance
                        field.update(
                            PdfDict(AP='{}'.format(data_dict[field_name]))
                        )

    # Save the filled form
    PdfWriter().write(output_pdf_path, template_pdf)

# Fill the form and save it
fill_pdf(pdf_template_path, output_pdf_path, form_data)

print(f"Formular erfolgreich ausgefüllt und gespeichert unter: {output_pdf_path}")
```
